```python
df = pd.read_excel("South Australia housing prices.xlsx")
display(df)
```

```python
df.dtypes
```

```python
df['Sales Q1 2022'] = df['Sales 1Q 2022'].fillna(0).astype(int)
df['Median Q1 2022'] = df['Median 1Q 2022'].fillna(0).astype(int)
df['Sales Q1 2023'] = df['Sales 1Q 2023'].fillna(0).astype(int)
df['Median Q1 2023'] = df['Median 1Q 2023'].fillna(0).astype(int)
df['Median Change'] = df['Median\nChange'].fillna(0).round(2)
columns_to_drop = ['Sales 1Q 2022', 'Median 1Q 2022', 'Sales 1Q 2023', 'Median 1Q 2023', 'Median\nChange']
df = df.drop(columns_to_drop, axis=1)
display(df)
```

```python
df.to_csv("South Australia housing prices Q4 2022 & Q1 2023 - clean version")
```
